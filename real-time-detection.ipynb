```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import joblib  # For loading the label dictionary

# Load the trained model
model = load_model('face_recognition_model.keras')

# Load the label dictionary
label_dict = joblib.load('label_dict.pkl')

# Set parameters
image_size = (128, 128)

# Real-time image classification
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame
    img = cv2.resize(frame, image_size)
    img = img / 255.0
    img = np.expand_dims(img, axis=0)

    # Predict the class
    predictions = model.predict(img)
    class_idx = np.argmax(predictions)
    class_label = list(label_dict.keys())[list(label_dict.values()).index(class_idx)]
    confidence = predictions[0][class_idx]

    # Display the result
    cv2.putText(frame, f"{class_label}: {confidence:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    cv2.imshow("Real-time Classification", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
